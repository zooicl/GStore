# Null Rate

```python
for c in df_train:
    null_rate = df_train[df_train[c].isnull()].shape[0] / df_train.shape[0]
    if null_rate > 0:
        print(c, null_rate)
```

# Value Counts

```python
for c in df_train:
    if is_object_dtype(df_train[c]):
        print('<{}>'.format(c))
        print(df_train[c].value_counts())
        print()
```
